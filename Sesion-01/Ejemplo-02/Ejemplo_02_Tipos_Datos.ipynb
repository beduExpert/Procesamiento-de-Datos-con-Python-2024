## 1. N칰meros
Python soporta diferentes tipos de n칰meros. Usaremos tres tipos principales: `int`, `float`, y `complex`.

### Enteros (`int`)
Los enteros son n칰meros sin parte decimal. Pueden ser positivos o negativos.

```python
# Declaramos una variable de tipo entero
x = 5
print(x, type(x))
```

### N칰meros de punto flotante (`float`)
Los flotantes representan n칰meros reales; incluyen parte decimal o son escritos en notaci칩n exponencial.

```python
# Declaramos una variable de tipo flotante.
y = 3.14
print(y, type(y))
```

### N칰meros complejos (`complex`)
Los n칰meros complejos tienen una parte real y una parte imaginaria.

```python
# Declaramos una variable de tipo complejo.
z = 3 + 4j
print(z, type(z))
```

## 2. Cadenas de caracteres (Strings)
Las cadenas son secuencias de caracteres utilizadas para almacenar texto.

```python
# Declaramos una variable de tipo texto.
cadena = "Hola, mundo!"
print(cadena, type(cadena))
```

## 3. Booleanos
Python soporta valores booleanos t칤picos, que son `True` y `False`.

```python
# Declaramos dos variables de tipo booleano.
verdadero = True
falso = False
print(verdadero, type(verdadero))
print(falso, type(falso))
```
